# Adding the target column in the dataset to predict the model with more accuracy

```python
iris_data=pd.read_csv(r'C:\Users\Dell\Downloads\archive (5)\IRIS.csv')
```

```python
iris_datafinal=pd.DataFrame(iris_data)
iris_datafinal['target']=0.0
```

```python
iris_datafinal
```

```python
target = []

for i in range(len(iris_datafinal['species'])):
    if iris_datafinal['species'][i] == 'Iris-setosa':
        target.append("0")
    elif iris_datafinal['species'][i] == 'Iris-versicolor':
        target.append("1")
        
    else:
        target.append("2")


iris_datafinal['target'] = target
```

```python
iris_datafinal.sample(10)
```

```python
iris_datafinal.tail()
```

```python
from sklearn.model_selection import train_test_split
X = iris_datafinal.drop(['target','species'], axis=1)
X = X.to_numpy()[:, (2,3)]
y = iris_datafinal['target']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5, random_state=42)
```

```python
X_test
```

```python
iristree = DecisionTreeClassifier()
iristree.fit(X_train, y_train)
ypred = iristree.predict(X_test)
```

```python
print(ypred)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(ypred, y_test))

print(classification_report(ypred, y_test))
```

```python
tree.plot_tree(iristree)
```
